```python
ratings_with_name=ratings.merge(book, on="ISBN")
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num-rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg-rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on="Book-Title")
popular_df
```

```python
top_50=popular_df[popular_df['num-rating']>=250].sort_values('avg-rating', ascending=False).head(50)
top_50=top_50.merge(book,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title", "Book-Author","Image-URL-M","num-rating","avg-rating"]].reset_index()
top_50
```
